````python
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
client = OpenAI(api_key=OPENAI_API_KEY)

def get_attractions_with_maps(destination, limit=3):
    prompt = f"""
    List the top {limit} tourist attractions in {destination}.
    For each attraction, provide:
      - name
      - a short description
      - a fun fact
      - Google Maps link (https://www.google.com/maps/search/<attraction>+{destination.replace(' ','+')})
    Respond ONLY as JSON array of objects with fields:
      name, description, fun_fact, map_link
    """
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}],
        temperature=0
    )
    content = response.choices[0].message.content.strip()
    if content.startswith("```"):
        content = content.strip("`").replace("json","").strip()
        if content.endswith("```"):
            content = content[:-3].strip()
    import json
    return json.loads(content)

if __name__ == "__main__":
    city = input("Enter destination city: ")
    attractions = get_attractions_with_maps(city, limit=3)
    print(f"\nTop {len(attractions)} attractions in {city}:\n")
    for i, a in enumerate(attractions,1):
        print(f"{i}. {a['name']}")
        print(f"   Description: {a['description']}")
        print(f"   Fun Fact: {a['fun_fact']}")
        print(f"   Map link: {a['map_link']}\n")
````
